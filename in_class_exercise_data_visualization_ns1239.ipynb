## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

try:
  penguins = sns.load_dataset('penguins')
  penguins = penguins.dropna()
except Exception:
  pass

plt.figure(figsize=(10,6))

sns.scatterplot(
    data = penguins,
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    hue = 'species'
)

plt.title('Penguin Bill Length vs. Bill Depth by Species')

plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')

plt.show()
```

**Your interpretation (2–3 sentences):**

_The Adelie Penguin overall appears to have a much longer bill depth, but a much shorter bill length compared to the other two species of penguin. The Chinstrap penguin lie more in the middle; they have an average bill depth and bill length with outliers leaning more towards the longer side of both. The Gentoo Penguin have an average bill length, leaning more towards the longer side, but a much shorter bill depth compared to the others._

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()

plt.figure(figsize=(10,5))

sns.histplot(
    data = penguins,
    x = 'bill_length_mm',
    bins = 10
)

plt.title('Penguin Bill Length (10 Bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')

plt.show()
```

```python
# TODO: Histogram with 30 bins

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()

plt.figure(figsize=(10,5))

sns.histplot(
    data = penguins,
    x = 'bill_length_mm',
    bins = 30
)

plt.title('Penguin Bill Length (30 Bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')

plt.show()
```

**Your comparison (1–2 sentences):**

_The 10 bin histogram gives a broad overview of the data and suggests a more symmetrical distribution of bill lengths. When increasing the number to 30 bins,we get a much more detailed structure with two peaks. This suggests there are two distinct groups of bill lengths that the 10 bin plot couldn't represent._

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()

plt.figure(figsize=(10,6))

sns.boxplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g'
)

plt.title('Penguin Body Mass Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')

plt.show()
```

**Your summary (1–2 sentences):**

_The boxplot clearly shows that Gentoo penguins are signficantly heavier on averange than the other two species, with the entire middle 50% of the data being almost above the Adelie and Chinstrap. The Adelie and Chinstrap species have similar lower body mass ranges, though the Chinstrap species are slightly lighter on average than the Adelie penguins._
